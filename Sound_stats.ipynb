### Частота дискретизации звука 
частота взятия отсчётов непрерывного по времени сигнала (Гц). С какой частотой записан звук, сколько раз в течении одной секунды осуществлялась запись сигнала (Например, когда мы говорим, что частота дискретизации 44,1 кГц, то это значит, что сигнал измеряется 44 100 раз в течение одной секунды) . Одно значение для всего аудиофайла

### Амплитуда звука 
График показывает величину амплитуды для каждой записи сигнала (Например, 44100 записей сигнала в секунду). Амплитуда показывает громкость звука.


```
plt.figure()
plt.plot(data)
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.show()
```

### Громкость звука в Децибелах. 
Децибелы высчитываются используя амплитуду звука по формуле e8e4645b.jpg


где p_rms - текущая амплитуда, взятая из графика, а p_ref - начальная амплитуда. Децибелы - относительная величина, а не абсолютная. И вычисляется она относительно p_ref 

```
p_ref = 0.5
dbs = np.copy(data)
for i in range(len(dbs)):
  p_rms = abs(int(dbs[i])) + 120
  dbs[i] = 20*math.log10(p_rms / p_ref)

plt.figure()
plt.plot(dbs)
plt.xlabel('Sample Index')
plt.ylabel('DB')
plt.show()
```
